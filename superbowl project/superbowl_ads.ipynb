```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
superbowl = pd.read_csv('/content/superbowl-ads.csv')
```

```python
superbowl.head(3)
```

```python
superbowl.columns
```

```python
superbowl.info()
```

```python
superbowl.drop(columns = ['superbowl_ads_dot_com_url','youtube_url'], inplace=True)
```

```python
superbowl["year"].unique()
```

```python
superbowl["brand"].unique()
```

```python
superbowl.sort_values(by='year', ascending = False, inplace=True)
```

```python
superbowl.head()
```

```python
superbowl['year'] = superbowl['year'].astype('string')
```

```python
Number_Ads_yearly = superbowl.groupby('year')['year'].value_counts().reset_index()
Number_Ads_yearly.sort_values(by='count', ascending = False, inplace=True)
```

```python
plt.figure(figsize=(12,4))
plt.bar(Number_Ads_yearly['year'], Number_Ads_yearly['count'], color='magenta')
plt.title("Total Number of ads per year")
plt.xlabel('Year')
plt.ylabel('Total Number of Ads')
plt.show()
```

```python
ys = superbowl.groupby('year')['year'].value_counts().reset_index()
ys.sort_values(by='year', ascending = True, inplace=True)
```

```python
plt.figure(figsize=(15, 4))
plt.plot(ys['year'], ys['count'], color='green')
plt.title('Number of Ads Yearly')
plt.xlabel('Year')
plt.ylabel('Number of Ads')
plt.show()
```

```python
brand = superbowl.groupby('brand')['brand'].value_counts().reset_index()
brand.sort_values(by='count', ascending = True, inplace=True)
```

```python
plt.figure(figsize=(12,4))
plt.barh(brand['brand'], brand['count'], color=['firebrick', 'green', 'blue', 'black', 'red',
                    'purple', 'seagreen', 'skyblue', 'black', 'tomato'])
plt.title("Total Number of ads per brand")
plt.ylabel('Brand')
plt.xlabel('Number of Ads')
plt.show()
```
